Question5:- Create a list and tuple in Python, and demonstrate how attempting to change an element work differently for each.

```python
my_list = [1, 2, 3, 4, 5]
my_tuple = (10, 20, 30, 40, 50)

print("Original List:", my_list)
print("Original Tuple:", my_tuple)

# Demonstrating mutability with a list
print("\nAttempting to change an element in the list:")
my_list[2] = 99
print("Modified List:", my_list)

# Demonstrating immutability with a tuple
print("\nAttempting to change an element in the tuple:")
try:
    my_tuple[2] = 999
except TypeError as e:
    print(f"Error: {e}")

print("Tuple after attempted modification (remains unchanged):")
print(my_tuple)
```

Question6:-Write a function to demonstrate the behavior of mutable and immutable arguments.

```python
def modify_list_in_function(my_list_arg):
    print(f"  Inside function: List before modification: {my_list_arg}")
    my_list_arg[0] = 100 # This will change the original list
    print(f"  Inside function: List after modification: {my_list_arg}")

def attempt_modify_tuple_in_function(my_tuple_arg):
    print(f"  Inside function: Tuple before attempted modification: {my_tuple_arg}")
    try:
        # This will raise a TypeError
        my_tuple_arg[0] = 1000
    except TypeError as e:
        print(f"  Inside function: Error trying to modify tuple: {e}")
    print(f"  Inside function: Tuple after attempted modification: {my_tuple_arg}")


# Demonstrating with a list (mutable)
my_original_list = [1, 2, 3]
print(f"Original list before function call: {my_original_list}")
modify_list_in_function(my_original_list)
print(f"Original list after function call: {my_original_list}\n")

# Demonstrating with a tuple (immutable)
my_original_tuple = (10, 20, 30)
print(f"Original tuple before function call: {my_original_tuple}")
attempt_modify_tuple_in_function(my_original_tuple)
print(f"Original tuple after function call: {my_original_tuple}")
```

Question7:-Write a program that performs basic arithmetic operation on tow user input numbers.

```python
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))

print(f"Addition: {num1} + {num2} = {num1 + num2}")
print(f"Subtraction: {num1} - {num2} = {num1 - num2}")
print(f"Multiplication: {num1} * {num2} = {num1 * num2}")

if num2 != 0:
    print(f"Division: {num1} / {num2} = {num1 / num2}")
else:
    print("Division by zero is not allowed.")
```

Question: Write a program to demonstrate the use of logical operators.

```python
x = 5
y = 10
z = 15

# Logical AND
print(f"Is x > 0 and y < 20? {x > 0 and y < 20}")  # True and True = True
print(f"Is x > 0 and z < 10? {x > 0 and z < 10}")  # True and False = False

# Logical OR
print(f"Is x < 0 or y < 20? {x < 0 or y < 20}")    # False or True = True
print(f"Is x < 0 or y < 5? {x < 0 or y < 5}")      # False or False = False

# Logical NOT
print(f"Is not (x == 5)? {not (x == 5)}")          # not True = False
print(f"Is not (y > z)? {not (y > z)}")            # not False = True
```
